```python
%%writefile app.py
import os
import io
import torch
import uvicorn
import spacy
import pdfplumber
import moviepy.editor as mp
import librosa
import soundfile as sf
import matplotlib.pyplot as plt
import numpy as np
from fastapi import FastAPI, UploadFile, File, HTTPException
from fastapi.responses import FileResponse
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer
from sentence_transformers import SentenceTransformer
from pyngrok import ngrok
from threading import Thread

# Ensure static directory exists
os.makedirs("static", exist_ok=True)

# ✅ Ensure GPU Usage
device = "cuda" if torch.cuda.is_available() else "cpu"

# ✅ Load NLP Models
try:
    if not spacy.util.is_package("en_core_web_sm"):
        spacy.cli.download("en_core_web_sm")

    nlp = spacy.load("en_core_web_sm")
    summarizer = pipeline("summarization", model="nsi319/legal-pegasus", device=0 if torch.cuda.is_available() else -1)
    embedding_model = SentenceTransformer("all-mpnet-base-v2", device=device)
    ner_model = pipeline("ner", model="dslim/bert-base-NER", tokenizer="dslim/bert-base-NER", device=0 if torch.cuda.is_available() else -1)
    speech_to_text = pipeline("automatic-speech-recognition", model="openai/whisper-base", device=0 if torch.cuda.is_available() else -1)

except Exception as e:
    raise RuntimeError(f"Error loading models: {str(e)}")

# ✅ Load Falcon 7B for Chatbot
MODEL_NAME = "tiiuae/falcon-7b-instruct"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
chatbot_model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.bfloat16,
    device_map="auto"
)

# ✅ Initialize FastAPI
app = FastAPI()

# ✅ PDF Text Extraction
def extract_text_from_pdf(pdf_file):
    """Extracts text from a PDF file using pdfplumber."""
    try:
        with pdfplumber.open(pdf_file) as pdf:
            text = "\n".join([page.extract_text() or "" for page in pdf.pages])
        if not text.strip():
            return None  # Return None if no readable text is found
        return text
    except Exception as e:
        raise HTTPException(status_code=400, detail=f"PDF extraction failed: {str(e)}")

# ✅ Legal Risk Assessment
def analyze_risk(text):
    """Analyzes legal risk in the document using keyword-based analysis."""
    risk_keywords = {
        "Liability": ["liability", "responsible", "responsibility", "legal obligation"],
        "Termination": ["termination", "breach", "contract end", "default"],
        "Indemnification": ["indemnification", "indemnify", "hold harmless", "compensate", "compensation"],
        "Payment Risk": ["payment", "terms", "reimbursement", "fee", "schedule", "invoice", "money"],
        "Insurance": ["insurance", "coverage", "policy", "claims"],
    }

    risk_scores = {category: 0 for category in risk_keywords}

    # Count keyword occurrences
    for category, keywords in risk_keywords.items():
        for keyword in keywords:
            risk_scores[category] += text.lower().count(keyword.lower())

    return risk_scores

# ✅ Generate Risk Assessment Bar Chart
def generate_risk_chart(risk_scores):
    """Generates a bar chart for risk assessment and saves it as an image."""
    categories = list(risk_scores.keys())
    values = list(risk_scores.values())

    plt.figure(figsize=(10, 6))
    bars = plt.barh(categories, values, color=['red', 'orange', 'yellow', 'blue', 'green'])

    # Add labels to the bars
    for bar in bars:
        width = bar.get_width()
        label_x_pos = width if width > 0 else 0.1
        plt.text(label_x_pos, bar.get_y() + bar.get_height()/2, f'{width}',
                va='center', ha='left', fontweight='bold')

    plt.xlabel("Risk Level (Keyword Mentions)")
    plt.title("Legal Risk Assessment")
    plt.tight_layout()

    # Save chart
    plt.savefig("static/risk_chart.png")
    plt.close()
    return "static/risk_chart.png"

# ✅ Extract Audio from Video
def extract_audio_from_video(video_path):
    """Extracts audio from a video file and saves it as a WAV file."""
    try:
        audio_path = f"{os.path.splitext(video_path)[0]}.wav"
        video = mp.VideoFileClip(video_path)
        video.audio.write_audiofile(audio_path, verbose=False, logger=None)
        return audio_path
    except Exception as e:
        print(f"Error extracting audio: {e}")
        return None

# ✅ Transcribe Audio
# ✅ Transcribe Audio (Fixed for Long Audio)
def transcribe_audio(audio_path):
    """Transcribes speech from an audio file using Whisper, handling long-form audio properly."""
    try:
        # Check if file exists
        if not os.path.exists(audio_path):
            return "⚠️ Audio file not found."

        # Load audio with librosa
        audio, sr = librosa.load(audio_path, sr=16000)
        duration = len(audio) / sr

        # ✅ If audio is longer than 30 seconds, split into smaller chunks
        if duration > 30:
            chunk_size = 30 * sr  # 30-second chunks
            chunks = [audio[i:i + chunk_size] for i in range(0, len(audio), chunk_size)]

            transcripts = []
            for idx, chunk in enumerate(chunks):
                temp_chunk_path = f"temp_chunk_{idx}.wav"
                sf.write(temp_chunk_path, chunk, sr)

                # ✅ Transcribe with Whisper
                result = speech_to_text(temp_chunk_path, return_timestamps=True)
                text = result.get("text", "").strip()

                if text:
                    transcripts.append(text)

                os.remove(temp_chunk_path)  # Clean up temp file

            return " ".join(transcripts) if transcripts else "⚠️ No speech detected in the audio."

        else:
            # ✅ Process normally if audio is <= 30 seconds
            result = speech_to_text(audio_path, return_timestamps=True)
            return result.get("text", "⚠️ No speech detected in the audio.")

    except Exception as e:
        return f"⚠️ Transcription error: {str(e)}"


# ✅ Legal Document Analysis API (Fixed)
@app.post("/analyze_legal_document")
async def analyze_legal_document(file: UploadFile = File(...)):
    """Analyzes a legal document by extracting text, summarizing, and identifying risks."""
    try:
        content = await file.read()
        text = extract_text_from_pdf(io.BytesIO(content))

        if not text:
            return {"status": "error", "message": "⚠️ No valid text found in the document."}

        # Debug print
        print(f"Extracted text length: {len(text)}")

        # Ensure proper summarization
        if len(text) < 100:
            summary = "Document is too short for meaningful summarization."
        else:
            # Process text in chunks if needed
            chunk = text[:4096] if len(text) > 4096 else text
            try:
                summary_result = summarizer(chunk, max_length=200, min_length=30, do_sample=False)
                summary = summary_result[0]['summary_text']
            except Exception as e:
                print(f"Summarization error: {e}")
                summary = "Error generating summary. The document might not contain standard text."

        # Ensure risk assessment runs
        risk_scores = analyze_risk(text)

        # Only generate chart if we have non-zero risk scores
        if any(risk_scores.values()):
            chart_path = generate_risk_chart(risk_scores)
            chart_url = "/download_risk_chart"
        else:
            chart_url = None

        return {
            "status": "success",
            "summary": summary,
            "risk_scores": risk_scores,
            "risk_chart": chart_url
        }

    except Exception as e:
        import traceback
        print(traceback.format_exc())
        return {"status": "error", "message": str(e)}

# ✅ Endpoint to Download Risk Chart
@app.get("/download_risk_chart")
async def download_risk_chart():
    if os.path.exists("static/risk_chart.png"):
        return FileResponse("static/risk_chart.png", media_type="image/png", filename="risk_chart.png")
    else:
        raise HTTPException(status_code=404, detail="Risk chart not found")

# ✅ Video Upload & Analysis API (Fixed)
@app.post("/analyze_video")
async def analyze_video(file: UploadFile = File(...)):
    """Extracts speech from video and analyzes it."""
    try:
        # Ensure file is a video
        if not file.filename.lower().endswith(('.mp4', '.avi', '.mov', '.mpeg4')):
            return {"status": "error", "message": "⚠️ Please upload a valid video file (.mp4, .avi, .mov, .mpeg4)"}

        # Save uploaded file
        video_path = f"temp_{file.filename}"
        with open(video_path, "wb") as f:
            f.write(await file.read())

        # Process the video
        try:
            audio_path = extract_audio_from_video(video_path)
            if not audio_path:
                return {"status": "error", "message": "⚠️ Failed to extract audio from video."}

            transcript = transcribe_audio(audio_path)

            # Clean up temp files
            try:
                os.remove(video_path)
                os.remove(audio_path)
            except:
                pass

            if not transcript or transcript.startswith("⚠️"):
                return {"status": "error", "message": transcript or "⚠️ No valid audio detected in the video."}

            return {"status": "success", "transcript": transcript}
        except Exception as e:
            import traceback
            print(f"Video analysis error: {traceback.format_exc()}")
            return {"status": "error", "message": f"⚠️ Error analyzing video: {str(e)}"}
    except Exception as e:
        return {"status": "error", "message": str(e)}

# ✅ Health Check Endpoint
@app.get("/health")
async def health_check():
    return {"status": "ok", "models_loaded": True}

# ✅ Restart Ngrok & Fix SSL Issue
def setup_ngrok():
    try:
        ngrok.kill()
        ngrok_tunnel = ngrok.connect(8500, "http")
        print(f"✅ NEW PUBLIC URL: {ngrok_tunnel.public_url}")
        return ngrok_tunnel
    except Exception as e:
        print(f"⚠️ Ngrok error: {e}")
        return None

def run():
    uvicorn.run(app, host="0.0.0.0", port=8500, timeout_keep_alive=300)

if __name__ == "__main__":
    ngrok_tunnel = setup_ngrok()
    Thread(target=run).start()


```
